```python
import pandas as pd
import json
import os              
```

```python
##reading all 2018 files and joining them in a single dataset
data2018 = []
for file in os.listdir("./Resources/Data1/2018"):
        data2018.append(pd.read_csv(f"./Resources/Data1/2018/{file}"))
df18 = pd.concat(data2018)
```

```python
#reading all 2019 files and joining them in a single dataset
data2019 = []
for file in os.listdir("./Resources/Data1/2019"):
        data2019.append(pd.read_csv(f"./Resources/Data1/2019/{file}"))
df19 = pd.concat(data2019)
```

```python
#reading all 2020 files and joining them in a single dataset
data2020 = []
for file in os.listdir("./Resources/Data1/2020"):
        data2020.append(pd.read_csv(f"./Resources/Data1/2020/{file}"))
df20 = pd.concat(data2020)
```

```python
#reading all 2021 files and joining them in a single dataset (excluding Jan data due to inconcistency)
data2021 = []
for file in os.listdir("./Resources/Data1/2021"):
        data2021.append(pd.read_csv(f"./Resources/Data1/2021/{file}"))
df21 = pd.concat(data2021)
```

```python
##grouping 2018 data
df18["Month"] = pd.to_datetime(df18['starttime']).dt.strftime('%Y-%m')
```

```python
df18_group = df18.groupby(["Month", "usertype", "gender","birth year"]).count()
```

```python
#grouping 2019 data
df19["Month"] = pd.to_datetime(df19['starttime']).dt.strftime('%Y-%m')
df19_group = df19.groupby(["Month", "usertype", "gender","birth year"]).count()
```

```python
##grouping 2020 data
df20["Month"] = pd.to_datetime(df20['starttime']).dt.strftime('%Y-%m')
df20_group = df20.groupby(["Month", "usertype", "gender","birth year"]).count()
```

```python
#joing all datasets for 2018-2020
df_concat = pd.concat([df18_group, df19_group, df20_group])
```

```python
#reset index
df_concat.reset_index(inplace = True)
df_concat
```

```python
#drop unused columns
df_concat_clean = df_concat[["Month", "usertype","gender","birth year", "tripduration"]]
df_concat_clean["Month"]= pd.to_datetime(df_concat_clean["Month"])
df_concat_clean
```

```python
#save output to csv
df_concat_clean.to_csv('./Resources/Data1/df_concat.csv', index = False)
```

```python
#adding trip duration
df21["Trip_Duration"] = (pd.to_datetime(df21["ended_at"])-pd.to_datetime(df21["started_at"])).dt.total_seconds()
df21
```

```python
#save 2021 output to csv
df21.to_csv('./Resources/Data1/df2021.csv', index = False)
```
